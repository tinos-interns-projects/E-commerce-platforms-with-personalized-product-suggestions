```python
import sqlite3
import uuid
from datetime import datetime, timedelta

def place_order(user_id, delivery_details, payment_method, items):
    conn = sqlite3.connect("db.sqlite")
    cursor = conn.cursor()

    order_id = str(uuid.uuid4())
    order_date = datetime.now().isoformat()
    delivery_date = (datetime.now() + timedelta(days=7)).isoformat()

    platform_fee = 5
    delivery_charge = 30
    total = 0

    for item in items:
        subtotal = item['price'] * item['quantity']
        total += subtotal + platform_fee

    if total >= 1000:
        delivery_charge = 0

    grand_total = total + delivery_charge

    # Insert into orders
    cursor.execute('''
        INSERT INTO orders (
            id, user_id, delivery_details, order_date, delivery_date,
            total_amount, delivery_charge, payment_method, status
        ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', (
        order_id, user_id, delivery_details, order_date, delivery_date,
        grand_total, delivery_charge, payment_method, 'Processing'
    ))

    # Insert into order_items
    for item in items:
        cursor.execute('''
            INSERT INTO order_items (
                order_id, product_id, quantity, price, img
            ) VALUES (?, ?, ?, ?, ?)
        ''', (
            order_id, item['product_id'], item['quantity'], item['price'], item['img']
        ))

    conn.commit()
    conn.close()
    return order_id
```

```python
import sqlite3
import random

# Connect to the SQLite database
conn = sqlite3.connect('db.sqlite')
cursor = conn.cursor()

# Fetch all product IDs
cursor.execute("SELECT id FROM Product")
products = cursor.fetchall()

# Update each product with a random offer value between 5 and 90
for (product_id,) in products:
    offer_value = random.randint(5,30)
    cursor.execute("UPDATE Product SET offer = ? WHERE id = ?", (offer_value, product_id))

# Commit changes and close connection
conn.commit()
conn.close()

print("✅ Random offer values (5-90%) added to all products.")
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('db.sqlite')
cursor = conn.cursor()

# Fetch all product IDs with their price and offer
cursor.execute("SELECT id, price, offer FROM Product")
products = cursor.fetchall()

# Update discounted_price for each product
for product_id, price, offer in products:
    if price and offer:
        discount = (offer / 100) * price
        discounted_price = round(price - discount, 2)
    else:
        discounted_price = price  # Fallback if price/offer is missing or 0

    cursor.execute("UPDATE Product SET discounted_price = ? WHERE id = ?", (discounted_price, product_id))

# Commit and close
conn.commit()
conn.close()

print("✅ discounted_price updated for all products based on price and offer.")
```
